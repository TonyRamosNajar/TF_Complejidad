```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import csv

def buscar_coincidencias():
    distrito1_busqueda = entrada_distrito1.get()
    distrito2_busqueda = entrada_distrito2.get()

    ruta_dataset = r"C:\Users\Tony\OneDrive\Escritorio\Ciclo2023-1\ComplejiAlgo\TF_RAMOS_CAUSSO\dataset.csv"  # Reemplaza con la ruta de tu archivo CSV

    coincidencias = []

    try:
        with open(ruta_dataset, 'r') as archivo_csv:
            lector_csv = csv.DictReader(archivo_csv, delimiter=';')
            next(lector_csv)  # Saltar la primera línea del archivo CSV (encabezados)

            for fila in lector_csv:
                distrito1 = fila['Distrito']
                distrito2 = fila['Llegada']
                aplicacion = fila['Partida']
                precio = fila['Precio']

                if distrito1.lower() == distrito1_busqueda.lower() and distrito2.lower() == distrito2_busqueda.lower():
                    coincidencias.append((distrito1, distrito2, aplicacion, precio))

            if coincidencias:
                mostrar_coincidencias(coincidencias)
            else:
                messagebox.showinfo("Coincidencias", "No se encontraron coincidencias.")

    except FileNotFoundError:
        messagebox.showerror("Error", "No se encontró el archivo CSV en la ruta especificada.")

def mostrar_coincidencias(coincidencias):
    global ventana_coincidencias
    ventana_coincidencias = tk.Toplevel(ventana)
    ventana_coincidencias.title("Coincidencias")
    ventana_coincidencias.geometry("600x400")

    etiqueta_coincidencias = tk.Label(ventana_coincidencias, text="Coincidencias encontradas:")
    etiqueta_coincidencias.pack()

    frame_coincidencias = ttk.Treeview(ventana_coincidencias)
    frame_coincidencias['columns'] = ('Distrito Inicial', 'Distrito Final', 'Aplicación Disponible', 'Precio')
    frame_coincidencias.column('#0', width=0, stretch=tk.NO)
    frame_coincidencias.column('Distrito Inicial', width=150)
    frame_coincidencias.column('Distrito Final', width=150)
    frame_coincidencias.column('Aplicación Disponible', width=150)
    frame_coincidencias.column('Precio', width=150)

    frame_coincidencias.heading('#0', text='', anchor=tk.W)
    frame_coincidencias.heading('Distrito Inicial', text='Distrito Inicial', anchor=tk.W)
    frame_coincidencias.heading('Distrito Final', text='Distrito Final', anchor=tk.W)
    frame_coincidencias.heading('Aplicación Disponible', text='Aplicación Disponible', anchor=tk.W)
    frame_coincidencias.heading('Precio', text='Precio', anchor=tk.W)

    for i, (distrito, distrito2, aplicacion, precio) in enumerate(coincidencias):
        frame_coincidencias.insert(parent='', index='end', iid=i, values=(distrito, distrito2, aplicacion, precio))

    frame_coincidencias.pack()

    boton_ver_detalle = tk.Button(ventana_coincidencias, text="Ver detalle", command=lambda: ver_detalle(frame_coincidencias))
    boton_ver_detalle.pack()

    boton_eliminar = tk.Button(ventana_coincidencias, text="Eliminar aplicación", command=lambda: eliminar_aplicacion(frame_coincidencias))
    boton_eliminar.pack()

def ver_detalle(frame_coincidencias):
    seleccionado = frame_coincidencias.focus()  # Obtener el índice del item seleccionado

    if seleccionado:
        distrito_inicial = frame_coincidencias.item(seleccionado)['values'][0]  # Obtener el valor de la columna 'Distrito Inicial'
        distrito_final = frame_coincidencias.item(seleccionado)['values'][1]  # Obtener el valor de la columna 'Distrito Final'
        aplicacion = frame_coincidencias.item(seleccionado)['values'][2]  # Obtener el valor de la columna 'Aplicación Disponible'
        precio = frame_coincidencias.item(seleccionado)['values'][3]  # Obtener el valor de la columna 'Precio'

        ventana_detalle = tk.Toplevel(ventana)
        ventana_detalle.title("Detalle de Coincidencia")
        ventana_detalle.geometry("400x150")

        etiqueta_mensaje = tk.Label(ventana_detalle, text="Detalles de la opción seleccionada:")
        etiqueta_mensaje.pack()

        etiqueta_distrito_inicial = tk.Label(ventana_detalle, text=f"Distrito Inicial: {distrito_inicial}")
        etiqueta_distrito_inicial.pack()

        etiqueta_distrito_final = tk.Label(ventana_detalle, text=f"Distrito Final: {distrito_final}")
        etiqueta_distrito_final.pack()

        etiqueta_aplicacion = tk.Label(ventana_detalle, text=f"Aplicación Disponible: {aplicacion}")
        etiqueta_aplicacion.pack()

        etiqueta_precio = tk.Label(ventana_detalle, text=f"Precio: {precio}")
        etiqueta_precio.pack()

        boton_de_acuerdo = tk.Button(ventana_detalle, text="De acuerdo", command=lambda: mostrar_agradecimiento(ventana_detalle))
        boton_de_acuerdo.pack()

        boton_rechazar = tk.Button(ventana_detalle, text="Rechazar", command=ventana_detalle.destroy)
        boton_rechazar.pack()

    else:
        # Mostrar mensaje de error si no se selecciona ningún item
        messagebox.showerror("Error", "No se ha seleccionado ninguna coincidencia.")

def eliminar_aplicacion(frame_coincidencias):
    seleccionado = frame_coincidencias.focus()  # Obtener el índice del item seleccionado

    if seleccionado:
        respuesta = messagebox.askyesno("Confirmar eliminación", "¿Estás seguro de que deseas eliminar la aplicación seleccionada?")
        if respuesta == tk.YES:
            frame_coincidencias.delete(seleccionado)
            messagebox.showinfo("Eliminación exitosa", "La aplicación ha sido eliminada.")
    else:
        # Mostrar mensaje de error si no se selecciona ningún item
        messagebox.showerror("Error", "No se ha seleccionado ninguna coincidencia.")

def mostrar_agradecimiento(ventana_anterior):
    global ventana_agradecimiento
    ventana_anterior.destroy()

    ventana_agradecimiento = tk.Toplevel(ventana)
    ventana_agradecimiento.title("Gracias por tu elección")
    ventana_agradecimiento.geometry("300x100")

    etiqueta_agradecimiento = tk.Label(ventana_agradecimiento, text="¡Gracias por tu elección!")
    etiqueta_agradecimiento.pack()

    boton_volver = tk.Button(ventana_agradecimiento, text="Seguir Explorando", command=ventana_agradecimiento.destroy)
    boton_volver.pack()

    boton_cerrar_todas = tk.Button(ventana_agradecimiento, text="Cerrar todas las ventanas", command=cerrar_todas_las_ventanas)
    boton_cerrar_todas.pack()

def cerrar_todas_las_ventanas():
    ventana_agradecimiento.destroy()
    ventana_coincidencias.destroy()
    ventana.destroy()

def mostrar_districs():
    global ventana_distritos
    ventana_distritos = tk.Toplevel(ventana)
    ventana_distritos.title("Distritos")
    ventana_distritos.geometry("200x550")

    distritos = ['Ancon', 'Barranco', 'Chorrillos', 'Comas', 'El Agustino', 
    'Jesus Maria', 'La Victoria', 'Lima', 'Lince', 'Los Olivos', 'Lurin', 
    'Miraflores', 'Pueblo Libre', 'Puente Piedra', 'Punta Negra', 'Rimac', 
    'San Borja', 'San Marin de Porres', 'San Miguel', 'Santa Anita', 
    'Santiago de Surco', 'SJL', 'Villa el Salvador', 'Villa Maria del Triunfo']

    for distrito in distritos:
        etiqueta_distrito = tk.Label(ventana_distritos, text=distrito)
        etiqueta_distrito.pack()


def seleccionar_distrito(ventana_distritos):
    distrito_seleccionado = ventana_distritos.focus()
    if distrito_seleccionado:
        distrito = ventana_distritos.item(distrito_seleccionado)['text']
        entrada_distrito1.delete(0, tk.END)
        entrada_distrito1.insert(0, distrito)
        ventana_distritos.destroy()

ventana = tk.Tk()

ventana.title("Búsqueda de Coincidencias por Distrito")
ventana.geometry("350x150")

etiqueta_distrito1 = tk.Label(ventana, text="Distrito 1:")
etiqueta_distrito1.pack()

entrada_distrito1 = tk.Entry(ventana)
entrada_distrito1.pack()

etiqueta_distrito2 = tk.Label(ventana, text="Distrito 2:")
etiqueta_distrito2.pack()

entrada_distrito2 = tk.Entry(ventana)
entrada_distrito2.pack()

boton_distritos_predefinidos = tk.Button(ventana, text="Distritos disponibles en nuestra app", command=mostrar_districs)
boton_distritos_predefinidos.pack()

boton_buscar = tk.Button(ventana, text="Buscar coincidencias", command=buscar_coincidencias)
boton_buscar.pack()

ventana.mainloop()
```
